```python
# list of imports
import numpy as np

import matplotlib.pyplot as plt
import cartopy.crs as ccrs
from shapely.geometry import Point

import cartopy.feature as cfeature
```

```python
# first plot
np.random.seed(1)
x = 360 * np.random.rand(100)
y = 180 * np.random.rand(100)-90

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(x,y)
```

```python
# map plot
lat = x
lon = y

to_lon = -79.3983
to_lat = 43.6609

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(1,1,1, projection=ccrs.AzimuthalEquidistant(central_longitude=to_lon,
    central_latitude=to_lat))
ax.scatter(lon, lat, transform=ccrs.RotatedPole())
ax.coastlines()
ax.stock_img()
```

```python
# Plot canada with Features
canada_east = -63
canada_west = -123
canada_north = 75
canada_south = 37


standard_parallels = (49,77)
central_longitude = -(91+52/60)

fig = plt.figure(figsize=(13,13))
ax = fig.add_subplot(1,1,1, projection=ccrs.LambertConformal(central_longitude=central_longitude, standard_parallels=standard_parallels))
ax.set_extent([canada_west,canada_east,canada_south,canada_north])
# ax.stock_img() # removed to use better resolution features
# ax.coastlines()
ax.gridlines()

# scale and resolution
land_50m = cfeature.NaturalEarthFeature('physical', 'land', '10m', edgecolor='k',facecolor=cfeature.COLORS['land'])
ax.add_feature(land_50m)

ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.LAKES)
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.RIVERS)
```
