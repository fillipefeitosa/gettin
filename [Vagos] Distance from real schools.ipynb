```python
# List of Imports
import geopandas as gpd
import geopy.distance
import geojson
```

```python
# Load data
vagos = gpd.read_file('data_gettin/vagos.geojson')
vagosCentroids = vagos.centroid

schools = gpd.read_file('data_gettin/schoolsUpdatedVagos/equipamentos_escolares_1_ciclo_2014.json')
schools.to_crs(epsg=4326)
```

```python
# Convert Data to WGS8
def convertCoordinates(geojson) {
    b = 20037508.34;
    for point in geojson:
        lon = point.x
        lat = point.y
        geojson['geometry']
    geojson.features.forEach(elt => {
    let lon = elt.geometry.coordinates[0];
    let lat = elt.geometry.coordinates[1];
    elt.geometry.coordinates[0] = lon * 180 / b;
    elt.geometry.coordinates[1] = Math.atan(Math.exp(lat * Math.PI / b)) * 360 / Math.PI - 90;
  });
}
```

```python
# Cost-Distance calculation
def distance(pointA, pointB):
    coordA = (pointA.x, pointA.y)
    coordB = (pointB.x, pointB.y)
    return geopy.distance.geodesic(coordA, coordB)

distanceSchoolCentroid = {}
for school in schools['geometry']:
    for centroid in vagosCentroids:
        distanceSchoolCentroid[school, centroid] = distance(school, centroid)

        
```
