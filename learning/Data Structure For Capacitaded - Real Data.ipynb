#### Calculo de Escolas
Custo retirado de [link]( https://www.cmjornal.pt/sociedade/detalhe/portugal_gasta_6200_eurosano_por_aluno_no_ensino_publico)

```python
# Classes to support modular structure
# Deprecated. Used a Dict of Lists instead. Would be a pain to instantiate all the necessary objects
class Location():
    def __init__(self, bgri):
        self.bgri = bgri
    coordinates = {}
    
class School(Location):
    def __init__(self, bgri, capacity=80):
        super().__init__(bgri)
        # cost in Euros per Year. Check link above.
        self.cost = capacity*6200
        self.capacity = capacity
    
class SubSection(Location):
    def __init__(self, bgri, demand):
        super().__init__(bgri)
        self.demand = demand
        
```

```python
tupleOfCapacity = (80, 120, 200)

dictList = defaultdict(list)

for bgri in list(listOfSubsections_2010.keys()):
    for capacity in tupleOfCapacity:
        cost = capacity*6200
        tempList = (capacity, cost)
        dictList[bgri].append(tempList)
        
for escola in dictList:
    for escolaTamanho in range(len(dictList[escola])):
        print(dictList[escola][escolaTamanho][1])

    
```
