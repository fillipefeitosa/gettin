### Loading GeoJson File
using geojson package [Link](https://stackoverflow.com/questions/42753745/how-can-i-parse-geojson-with-python)


```python
import geojson
with open('../data_shpTest/states.geojson') as f:
    gj = geojson.load(f)
features = gj['features'][0]
print(json.dumps(features, indent=2))
```
